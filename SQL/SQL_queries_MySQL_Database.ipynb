**Find the average payment amount in each year and month, for which avergae payment is greater than 40000, from the table 'payments'. List in ascending order of the year**

```python
%%sql 
SELECT YEAR(paymentDate) AS Year, MONTH(paymentDate) AS Month , AVG(amount) AS Amount
FROM payments 
GROUP BY YEAR(paymentDate), MONTH(paymentDate)
HAVING Amount > 40000
ORDER BY YEAR(paymentDate), Amount DESC ;
```

**Get the no. of orders of each customer grouped by month of order joining the two tables - 'customers' and 'orders' using INNER JOIN**

```python
%%sql 
SELECT c.customerNumber, MONTH(o.orderDate) AS Month, COUNT(*) AS count 
FROM customers AS c, orders AS o 
WHERE c.customerNumber = o.customerNumber 
GROUP BY c.customerNumber, Month 
HAVING count > 1
ORDER BY count ;
```

**Writing the above query using a traditional syntax for INNER JOIN**

```python
%%sql 
SELECT c.customerNumber, MONTH(o.orderDate) AS Month, COUNT(*) AS count 
FROM customers AS c INNER JOIN orders AS o 
ON c.customerNumber = o.customerNumber 
GROUP BY c.customerNumber, Month 
HAVING count > 1
ORDER BY count ;
```

**example for LEFT JOIN :  tables 'products'-'orderdetails'**

```python
%%sql 
SELECT p.productCode, p.productLine, p.buyPrice, o.quantityOrdered, o.priceEach, o.orderLineNumber
FROM products AS p LEFT JOIN orderdetails AS o 
ON p.productCode = o.productCode
WHERE p.productLine = 'Motorcycles' AND p.buyPrice > 90
```

**The same query as above using RIGHT JOIN**

```python
%%sql 
SELECT p.productCode, p.productLine, p.buyPrice, o.quantityOrdered, o.priceEach, o.orderLineNumber
FROM orderdetails AS o RIGHT JOIN  products AS p  
ON p.productCode = o.productCode
WHERE p.productLine = 'Motorcycles' AND p.buyPrice > 90
```
